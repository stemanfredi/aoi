# AOI — interactive coastal sensor-siting chart

The Leaflet map IS the AOI: frame by pan/zoom, drop sensors and
measurements, hit **Preview bathy** to see the actual chart underneath
your markers, then **Save PNG** for the deliverable.

**Workflow**
1. Pick an aspect ratio (Map tab, top of the controls panel).
2. Pan / zoom the map to your area.
3. **Preview bathy** — fetches data once, overlays the rendered chart
   on the Leaflet map at the captured bounds. Place sensors and
   measurements on top of the actual bathy.
4. **Save PNG** — re-renders at full resolution with all your sensors +
   measurements + title. Reuses the previewed cache, so it's fast.

Filenames embed lat/lon + timestamp (`<lat>N_<lon>E_<YYYYMMDDTHHMMSS>.png`)
so renders never overwrite.

All logic lives in `designer.py`. `%run` reloads everything fresh each
cell execution — edit the file, re-run, see changes.
